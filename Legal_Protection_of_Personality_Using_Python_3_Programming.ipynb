1. **Анонимизация персональных данных**
Пример кода для хеширования персональных данных (например, имени и email) с помощью SHA-256:

```python
import pandas as pd
from hashlib import sha256

def anonymize_data(df, columns):
    for col in columns:
        df[col] = df[col].apply(lambda x: sha256(x.encode('utf-8')).hexdigest())
    return df

# Примерные данные
data = pd.DataFrame({
    'name': ['Иван Иванов', 'Мария Петрова', 'Алексей Смирнов'],
    'email': ['ivan@mail.ru', 'maria@mail.ru', 'alex@mail.ru'],
    'age': [30, 25, 40]
})

anonymized = anonymize_data(data, ['name', 'email'])
print(anonymized)
```

**Комментарий:**

Хеширование обеспечивает невозможность восстановления исходных данных, что соответствует принципам минимизации и безопасности обработки персональных данных по GDPR.

2. **Визуализация правовых рисков и инцидентов**

Для мониторинга нарушений можно использовать библиотеки matplotlib и seaborn. Пример визуализации роста числа инцидентов с утечками данных:

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = pd.DataFrame({
    'Год': [2021, 2022, 2023, 2024],
    'Утечки данных': [50, 75, 120, 180]
})

plt.figure(figsize=(8,5))
sns.lineplot(data=data, x='Год', y='Утечки данных', marker='o')
plt.title('Динамика утечек персональных данных (2021-2024)')
plt.xlabel('Год')
plt.ylabel('Количество инцидентов')
plt.grid(True)
plt.show()
```

Данная статья демонстрирует, как программирование на Python 3 может стать эффективным инструментом правовой защиты личности в эпоху ИИ, сочетая технические решения с соблюдением актуального законодательства.

**Комментарий:**

Визуализация помогает выявлять тенденции и своевременно принимать меры по усилению защиты.
